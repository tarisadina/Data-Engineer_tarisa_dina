```python
import pandas as pd

# Mengganti backslash dengan slash miring ke depan
data_bakery = pd.read_csv('C:/Users/hello/OneDrive/kampus-merdeka_data-engineer/21_Data Visualization/Pratikum/bakery.csv')
data_bakery
```

```python
import matplotlib.pyplot as plt

# Menghitung jumlah nilai transaksi per hari
daily_transaction_amount = data_bakery.groupby('transaction_date')['transaction_amount'].sum()

# Membuat line plot
plt.figure(figsize=(10, 6))
plt.plot(daily_transaction_amount.index, daily_transaction_amount.values, marker='o')
plt.title('Jumlah Nilai Transaksi per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Nilai Transaksi (Rp)')
plt.grid(True)
plt
```

```python
import pandas as pd

# Membaca data dari file CSV
data_bakery = pd.read_csv('C:/Users/hello/OneDrive/kampus-merdeka_data-engineer/21_Data Visualization/Pratikum/bakery.csv')

# Mengonversi kolom 'transaction_date' menjadi tipe data datetime dengan format yang sesuai
data_bakery['transaction_date'] = pd.to_datetime(data_bakery['transaction_date'], format='%d-%m-%Y')

# Menghitung rata-rata nilai rating pelanggan per hari
rata2_rating_per_hari = data_bakery.groupby(data_bakery['transaction_date'].dt.date)['customer_rating'].mean()

# Menampilkan rata-rata nilai rating pelanggan per hari
print("\nRata-rata nilai rating pelanggan per hari:")
print(rata2_rating_per_hari)
```
