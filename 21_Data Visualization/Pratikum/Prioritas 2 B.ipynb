```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Mengganti backslash dengan slash miring ke depan
survey = pd.read_csv('C:/Users/hello/OneDrive/kampus-merdeka_data-engineer/21_Data Visualization/Pratikum/survey.csv')
survey
```

```python
# Menghitung jumlah responden untuk setiap kategori umur
jumlah_responden_per_umur = survey['age'].value_counts()

# Menghitung persentase responden untuk setiap kategori umur
persentase_responden_per_umur = (jumlah_responden_per_umur / jumlah_responden_per_umur.sum()) * 100

# Menampilkan hasil
persentase_responden_per_umur
```

```python
import matplotlib.pyplot as plt

# Visualisasi: Pie Chart untuk persentase responden berdasarkan umur
plt.figure(figsize=(10, 7))
plt.pie(persentase_responden_per_umur, labels=persentase_responden_per_umur.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Persentase Responden Berdasarkan Umur')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Menghitung frekuensi penggunaan bahasa pemrograman/framework untuk pengembangan front-end
bahasa_framework_frontend = survey['favorite_front_end'].value_counts()
bahasa_framework_frontend
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi: Bar Plot untuk bahasa pemrograman/framework yang paling banyak digunakan dalam pengembangan front-end
plt.figure(figsize=(12, 6))
sns.barplot(x=bahasa_framework_frontend.index, y=bahasa_framework_frontend.values, palette='viridis')
plt.title('Bahasa Pemrograman/Framework yang Paling Banyak Digunakan dalam Pengembangan Front-End')
plt.xlabel('Bahasa Pemrograman/Framework')
plt.ylabel('Jumlah Pengguna')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Menghitung frekuensi penggunaan bahasa pemrograman/framework untuk pengembangan front-end
bahasa_framework_backend = survey['favorite_back_end'].value_counts()
bahasa_framework_backend
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi: Bar Plot untuk bahasa pemrograman/framework yang paling banyak digunakan dalam pengembangan front-end
plt.figure(figsize=(12, 6))
sns.barplot(x=bahasa_framework_backend.index, y=bahasa_framework_backend.values, palette='viridis')
plt.title('Bahasa Pemrograman/Framework yang Paling Banyak Digunakan dalam Pengembangan backend')
plt.xlabel('Bahasa Pemrograman/Framework')
plt.ylabel('Jumlah Pengguna')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
